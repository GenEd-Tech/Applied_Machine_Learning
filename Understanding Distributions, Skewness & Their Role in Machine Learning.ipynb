##1. Introduction to Distributions

A distribution in statistics is a way to describe how values of a variable are spread or clustered. (Shop Example)

1.1 Normal Distribution (Gaussian)

Bell-shaped curve

Mean = Median = Mode

Symmetrical about the center

Most real-world data follows this pattern (e.g., height, IQ scores)

1.2 Log-Normal Distribution

Skewed to the right

Occurs when data is positively skewed, but the log of the data is normally distributed

Common in economics (income, stock prices)

1.3 Positive Skewed Distribution

Tail on the right side

Mean > Median

Data has outliers or extreme high values

Example: Salary, property prices

1.4 Negative Skewed Distribution

Tail on the left side

Mean < Median

Data has extreme low values

Example: Exam scores when most students score high



##2. Mean, Standard Deviation & Skewness

2.1 Mean

Average value

Sensitive to outliers (e.g., one extremely high value pulls the mean up)

2.2 Standard Deviation (SD)

Measures spread or dispersion around the mean

A small SD means data is close to the mean; large SD means it's spread out

2.3 Skewness

Measures asymmetry of distribution

0 = perfectly symmetrical (normal)

Positive skew: tail right

Negative skew: tail left

##3. Impact on Machine Learning

Machine Learning models assume normally distributed data in many cases (e.g., Linear Regression, Logistic Regression).

Why Skewness Matters:

Skewed data may lead to biased predictions

Outliers can dominate the model

Feature scaling and normalization become harder

Solutions:

Log Transformation to reduce positive skew

Z-score normalization for standardization

Box-Cox or Power Transforms for fixing skewness

##4. The Joke: Don’t Drown in Averages!

"A man drowned in a river that was, on average, only 4 feet deep."

Scenario:

40m of river = 4 ft deep

10m of river = 12 ft deep

40m again = 4 ft deep

Average depth:



But in the middle, it's 12 ft. If the person can't swim, they still drown.

**Moral:**

Don’t blindly trust the mean. Always explore the shape of the data (use median, boxplots, histograms).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
normal_data = np.random.normal(loc=50, scale=10, size=1000)
positive_skew = np.random.exponential(scale=20, size=1000)
negative_skew = -1 * np.random.exponential(scale=20, size=1000) + 100
log_normal_data = np.random.lognormal(mean=3.5, sigma=0.5, size=1000)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.histplot(normal_data, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title("Normal Distribution")

sns.histplot(positive_skew, kde=True, ax=axes[0, 1], color='orange')
axes[0, 1].set_title("Positive Skewed")

sns.histplot(negative_skew, kde=True, ax=axes[1, 0], color='green')
axes[1, 0].set_title("Negative Skewed")

sns.histplot(log_normal_data, kde=True, ax=axes[1, 1], color='purple')
axes[1, 1].set_title("Log-Normal Distribution")

plt.tight_layout()
plt.show()
```

##Summary

Use mean and standard deviation to understand data spread

Use skewness and visual plots to detect data issues

Apply transformations when necessary before ML

Always explore the shape of the data, not just the average
